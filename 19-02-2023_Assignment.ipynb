**Q1. What is Flask Framework? What are the advantages of Flask Framework?**

Flask is a micro-web framework that is used to create web applications. It is considered microframework because it does not need any other libraries to work. It is designed to be easy to use to create a web applications quickly.  
Advantages of using Flask Framework are:-  
* It is lightweight i.e. it is easy to get running
* Flask is flexible and open to full control i.e. we can we can define modules and routes.
* It implements REST API between the flask server application and web application.

**Q3. What is App routing in Flask? Why do we use app routes?**

App routing in Flask is the process of creating a route using URL to functions which handle specific requests. We route to a specific functions which are requested. App routing can be done using 'app.route("/")'.  
App routes are used because it is easier to organise which functions are used for which webpage and makes navigating the web application easier.

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.**

There are several methods for URL Building in flask in Python. The most commonly used method is using 'url_for()' function. The 'url_for()' is a function in flask that is used to generate URLs dynamically for different rouctes in our flask app. The 'url_for()' function takes the name of the route or view as an argument and returns a URL that points to that route.
