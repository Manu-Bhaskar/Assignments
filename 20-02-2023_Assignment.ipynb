**Q1. Explain GET and POST methods.**

GET and POST are HTTP (Hyper Text Transfer Protocol) methods or verbs.  
HTTP is a communication protocol that is used to communicate between client and servers. It does so using differnt rules which are defined in different methods or verbs.  
GET is a HTTP method that is used to request data from client. Data is usually given in urls in name/value pairs.  
POST is HTTP method that is also used to request data from client side but in body of the template or any other form of data structure.
